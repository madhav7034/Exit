```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

def compute_vif(df_, features):
    X = df_[features].dropna()
    if X.shape[0] == 0:
        raise ValueError("No complete rows to compute VIF. Drop/clean NaNs first.")
    Xc = X.copy()
    for c in Xc.columns:
        if Xc[c].nunique() <= 1:
            Xc[c] = Xc[c] + np.random.normal(0, 1e-8, size=Xc.shape[0])
    vifs = []
    for i, f in enumerate(Xc.columns):
        try:
            v = variance_inflation_factor(Xc.values, i)
        except Exception:
            v = np.nan
        vifs.append((f, v))
    return pd.DataFrame(vifs, columns=['feature','VIF']).set_index('feature')

def reduce_by_vif_and_corr(df_, features, max_features=5, vif_thresh=5.0, corr_thresh=0.85):
    features = list(features)
    history = []
    while len(features) > max_features:
        try:
            vifs = compute_vif(df_, features)
        except Exception:
            vifs = pd.DataFrame({'VIF': [np.nan]*len(features)}, index=features)
        if vifs['VIF'].dropna().max() > vif_thresh:
            drop = vifs['VIF'].idxmax()
            reason = f"high VIF {vifs.loc[drop,'VIF']:.2f}"
            features.remove(drop)
            history.append((drop, reason))
            continue
        corr_sub = df_[features].corr().abs()
        np.fill_diagonal(corr_sub.values, 0)
        max_c = corr_sub.values.max()
        if max_c >= corr_thresh:
            i,j = np.unravel_index(np.argmax(corr_sub.values), corr_sub.shape)
            f1, f2 = corr_sub.index[i], corr_sub.columns[j]
            mean_abs = corr_sub.mean()
            drop = mean_abs.idxmax()
            reason = f"high corr {max_c:.2f} between {f1} and {f2}; dropped {drop}"
            features.remove(drop)
            history.append((drop, reason))
            continue
        mean_abs = df_[features].corr().abs().mean()
        drop = mean_abs.idxmax()
        reason = f"excess features; dropped {drop} (mean abs corr {mean_abs[drop]:.3f})"
        features.remove(drop)
        history.append((drop, reason))
    return features, history

selected, history = reduce_by_vif_and_corr(df, audio_cols, max_features=5, vif_thresh=5.0, corr_thresh=0.85)
print("Selected features (algorithm):", selected)
print("\nRemoval history:")
for f, r in history:
    print(f"- {f}: {r}")

try:
    print("\nFinal VIFs:")
    print(compute_vif(df, selected))
except Exception as e:
    print("Could not compute VIFs (not enough complete rows).", e)
```

```python

import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
import joblib

FEATURES = ['acousticness', 'speechiness', 'key', 'liveness', 'mode']
TARGET = 'Points'
IQR_MULTIPLIER = 1.5
RANDOM_STATE = 42

missing_cols = [c for c in FEATURES + [TARGET] if c not in df.columns]
if missing_cols:
    raise KeyError(f"Missing required columns in dataset: {missing_cols}")

df_model = df[FEATURES + [TARGET]].copy()
df_model[FEATURES] = df_model[FEATURES].apply(pd.to_numeric, errors='coerce')
df_model[TARGET] = pd.to_numeric(df_model[TARGET], errors='coerce')


df_model = df_model[df_model[TARGET].notna()].reset_index(drop=True)

cols_for_outlier = FEATURES + [TARGET]
bounds = {}
for c in cols_for_outlier:
    col = df_model[c].dropna()
    if col.empty:
        bounds[c] = (np.nan, np.nan)
        continue
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - IQR_MULTIPLIER * iqr
    upper = q3 + IQR_MULTIPLIER * iqr
    bounds[c] = (lower, upper)

outlier_mask = pd.Series(False, index=df_model.index)
for c, (low, up) in bounds.items():
    if np.isnan(low) or np.isnan(up):
        continue
    mask_c = df_model[c].notna() & ((df_model[c] < low) | (df_model[c] > up))
    outlier_mask = outlier_mask | mask_c


df_clean = df_model.loc[~outlier_mask].reset_index(drop=True)

if len(df_clean) < 20:
    raise RuntimeError("Too few rows after outlier removal. Reduce IQR multiplier or skip outlier removal.")


X = df_clean[FEATURES].copy()
y = df_clean[TARGET].copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[('num', numeric_pipeline, FEATURES)],
    remainder='drop',
    verbose_feature_names_out=False
)


gbr_pipeline = Pipeline([
    ('preproc', preprocessor),
    ('gbr', GradientBoostingRegressor(
        n_estimators=200,
        learning_rate=0.1,
        max_depth=3,
        random_state=RANDOM_STATE
    ))
])


gbr_pipeline.fit(X_train, y_train)


y_pred = gbr_pipeline.predict(X_test)
test_mae = mean_absolute_error(y_test, y_pred)

model_path = 'model_pipeline.pkl'
joblib.dump(gbr_pipeline, model_path)

print(f"{test_mae:.3f}")
```
